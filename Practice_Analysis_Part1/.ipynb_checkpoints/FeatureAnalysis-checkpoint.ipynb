```python
import numpy as np
import pandas as pd
import statistics
from sklearn import preprocessing
import statsmodels.api as sm
```

```python
train_df = pd.read_csv('https://grantmlong.com/data/SE_rents2018_train.csv', index_col=0)
test_set1_df = pd.read_csv('https://grantmlong.com/data/SE_rents2018_test1.csv', index_col=0)
test_set2_df = pd.read_csv('https://grantmlong.com/data/SE_rents2018_test2.csv', index_col=0)
```

```python
combine_df = pd.concat([train_df,test_set1_df,test_set2_df])
combine_df = combine_df.select_dtypes(include=[np.number])
combine_df = combine_df.drop(['bin', 'bbl','building_id'], axis=1)
combine_df.describe()
```

## Data Analysis Result 
1. Outliners - Replace value
   - bathrooms -> replace any row with number of bathrooms greater than 7 or equal to 0 with median
   - size_sqft -> replace any value greater than 8000 or equal to 0 with median
   - year_built -> replace any value equal to 0 with median
   - min_to_subway -> replace any value equal to 103343.6167 with median

```python
# Bedroom Analysis
print("0 Bedroom: ", combine_df.loc[combine_df.bedrooms==0].shape[0])
print("More than 6 Bedroom: ", combine_df.loc[combine_df.bedrooms>6].shape[0])
extreme_df = combine_df.loc[combine_df.bedrooms==0]
print(extreme_df[['rent','bedrooms']].head(10))
combine_df['bedrooms'].hist(bins=8)
extreme_df.describe()
```

```python
# Bathroom Analysis - Need to Drop Outliners
print("0 bathrooms: ", combine_df.loc[combine_df.bathrooms==0].shape[0])
extreme_df = combine_df.loc[combine_df.bathrooms>7]
print("More than 7 bathrooms: ", extreme_df.shape[0])
print(extreme_df[['rent','bathrooms']].head(10))
combine_df['bathrooms'].hist(bins=20)
```

```python
# size_sqft Analysis - Need to Fix Outliners
extreme_df = combine_df.loc[combine_df.size_sqft>6000]
print("extreme_df.shape: ", extreme_df.shape)
print(extreme_df[['rent','size_sqft']].head(10))
combine_df['size_sqft'].hist(bins=100)
```

```python
# size_sqft Analysis - Need to Drop Outliners
extreme_df = combine_df.loc[combine_df.size_sqft == 0]
print("extreme_df.shape: ", extreme_df.shape)
print(extreme_df[['rent','size_sqft']].head(10))
extreme_df.describe()
```

```python
# floor_count Analysis 
print("0 floor_count: ", combine_df.loc[combine_df.floor_count==0].shape[0])
extreme_df = combine_df.loc[combine_df.floor_count>80]
print("extreme_df.shape: ", extreme_df.shape)
print(extreme_df[['rent','floor_count']].head(10))
combine_df['floor_count'].hist(bins=100)
```

```python
# year_built Analysis - Replace year as 0 to different value
extreme_df = combine_df.loc[combine_df.year_built==0]
print("extreme_df.shape: ", extreme_df.shape)
print(extreme_df[['rent','year_built','has_elevator']].head(10))
combine_df['year_built'].hist(bins=50)
```

```python
# min_to_subway Analysis - Replace min_to_subway as 103343.6167 to different value
extreme_df = combine_df.loc[combine_df.min_to_subway== 103343.6167]
print("extreme_df.shape: ", extreme_df.shape)
print(extreme_df[['rent','min_to_subway','bathrooms','addr_zip']].head(10))
combine_df['min_to_subway'].hist(bins=100)
```

```python
# floornumber Analysis 
extreme_df = combine_df.loc[combine_df.floornumber> 60]
print("extreme_df.shape: ", extreme_df.shape)
print(extreme_df[['rent','floornumber']].head(10))
combine_df['floornumber'].hist(bins=100)
```
